Q9. Generate a NumPy array of 100 random integers between 0 and 1000. Find and display all prime numbers in
this array.

```python
array = np.random.randint(0, 1001, 100)
array
def is_prime(num):
    if num <= 1:
      return False
    if num == 2:
      return True
    if num % 2 == 0:
      return False
    for i in range(3, num, 2):
      if num % i == 0:
        return False
    return True

prime_numbers = []
for num in array:
  if is_prime(num):
    prime_numbers.append(num)
print(prime_numbers)
```
